## Amazon Scripting ##

```python

rawdata=rq.get("https://www.amazon.in/All-new-Echo-Dot-3rd-Gen/dp/B0792KTHKK/ref=sr_1_1?s=amazon-devices&ie=UTF8&qid=1553324424&sr=1-1&keywords=echo+dot")
data=rawdata.text
print(data)
```

```python
t1=data.find('id="productTitle"')
print(t1)
```

```python
raw1=data[t1+len('id="productTitle" class="a-size-large">'): ]
```

```python
end=raw1.find("</span>")
print(raw1[:end].strip())
```

```python
t2=data.find('id="priceblock_saleprice"')
print(t2)
```

```python
raw2=data[t2+len('id="priceblock_saleprice" '): ]
#print(raw2)
start=raw2.find('</span>')
print(raw2[start+len('</span>'):start+20-len('</span>')].strip())
```

```python
t3=data.find('class="a-icon-alt"')
print(t3)
```

```python
raw3=data[t3+len('class="a-icon-alt">'):]
t4=raw3.find('</span>')
print(raw3[0:t4])
```
