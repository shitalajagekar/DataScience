# WordCloud of Wikipedia

```python
import requests as rq
```

```python
raw_content=rq.get("https://en.wikipedia.org/wiki/Science")
```

```python
content=raw_content.text
```

```python
print(content)
```

```python
firstend=content.find("<script>")
print(firstend)
```

```python
lastend=content.find("</script>")
print(lastend)
```

```python
article_data=content[firstend+len("<script>"):lastend]
```

```python
print(article_data)
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopword=set(STOPWORDS)
```

```python
article=WordCloud(background_color="White", max_words=300, stopwords=stopword)
```

```python
article.generate(article_data)
```
