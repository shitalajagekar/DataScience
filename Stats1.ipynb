# GAUSSIAN DISTRIBUTION

```python
import numpy as np
import seaborn as sns
v=np.random.randint(low=1,high=7,size=10000) #more the size values will start having equal probability
print(v.mean())
plt.hist(v,bins=12)
```

Simulation of die rolls

```python
import numpy as np
import seaborn as sns
d1=np.random.randint(low=1,high=7,size=10000) 
d2=np.random.randint(low=1,high=7,size=10000)
avg=(np.array(d1)+np.array(d2))/2
print(avg.mean())
print(avg.var())
sns.distplot(avg,bins=12)
```

```python
d1=np.random.randint(low=1,high=7,size=10000) 
d2=np.random.randint(low=1,high=7,size=10000)
d3=np.random.randint(low=1,high=7,size=10000)
avg=(np.array(d1)+np.array(d2)+np.array(d3))/3
print(avg.mean())
print(avg.var())
sns.distplot(avg,bins=12)
```

```python
d1=np.random.randint(low=1,high=7,size=10000) 
d2=np.random.randint(low=1,high=7,size=10000)
d3=np.random.randint(low=1,high=7,size=10000)
d4=np.random.randint(low=1,high=7,size=10000)
avg=(np.array(d1)+np.array(d2)+np.array(d3)+np.array(d4))/4
print(avg.mean())
print(avg.var())
sns.distplot(avg,bins=12) ##resultant curve is known as gaussian or normal distribution
```

# Uniform DISTRIBUTION

```python
import seaborn as sns
from scipy.stats import uniform

data_uniform=uniform.rvs(size=10000,loc=10,scale=30)
sns.distplot(data_uniform,bins=100)
```

# Gamma DISTRIBUTION

```python
from scipy.stats import gamma

data_gama = gamma.rvs(size=10000,a=7)
val=np.array(data_gama)
sns.distplot(data_gama,bins=100)
```

# EXPONENTIAL DISTRIBUTION

```python
from scipy.stats import expon

data_expon = expon.rvs(size=10000,scale=1,loc=0)
val=np.array(data_expon)
sns.distplot(data_expon,bins=100)
```

# BERNOULLI DISTRIBUTION

```python
from scipy.stats import bernoulli

data_bernoulli = bernoulli.rvs(size=10000,p=0.6)
val=np.array(data_bernoulli)
sns.distplot(data_bernoulli,bins=100)
```

# BINOMIAL DISTRIBUTION
Frequency distrbution of the possible number of succesful outcomes in a given no. of trials in each of which there is same probabibility of success

```python
from scipy.stats import binom

data_binomial = binom.rvs(size=10000,p=0.8,n=10)
val=np.array(data_binomial)
sns.distplot(data_binomial,bins=100)
```

# POISSON DISTRIBUTION

```python
from scipy.stats import poisson

data_poisson = poisson.rvs(size=10000,mu=5)
val=np.array(data_poisson)
sns.distplot(data_poisson,bins=100)
```

# Asymptopia
Refers to the behaviour of estimators as the sample size goes to infinity

```python
observations = []
mean_values = []

for i in range(1,10000):
    v=np.random.randint(low=0,high=2,size=1)
    observations.append(i)
    mean_values.append(sum(v)/i)
plt.plot(observations,mean_values)
```
