```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\Admin\Documents\iris_csv.csv")
data.head(100)
```

```python
X=data.iloc[:,0:4].values
y=data.iloc[:,4].values
```

```python
df=pd.DataFrame(data=np.column_stack((data.data,data.target)),column=(*data.feature_names,"class"))
df.Class.replace([0,1,2],['setosa','versicolor','virginica'],inplace=true)
df
```

```python
pd.plotting.scatter_matrix(data,figsize=(15,10),diagonal='hist')
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.5,random_state=0)
```

```python
nb=GaussianNB()
model=nb.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
prediction=pd.DataFrame(np.column_stack((X_test,y_test,y_pred)),columns=["original class","predicted_class"])
prediction.original_Class.replace([0,1,2],[ 'setosa','versicolor','virginica'],inplace=true)
prediction.Predicted_Class.replace([0,1,2],[ 'setosa','versicolor','virginica'],inplace=true)
prediction
```

```python
accuracy=prediction.loc[prediction['Original_Class']==prediction['Original_class']].count()[0]/prediction.count()[0]
print(f'Accuracy of model={accuracy*100}%')
```
